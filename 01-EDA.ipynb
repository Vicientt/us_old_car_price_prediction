## This is a very dirty dataset with so null value for sum column, we're gonna explore each variable and think logically about what approaches we should take with each variable

#### 1. id

```python
# This variable does not relate anything to car/ car price as it just the id that the crawler take from a website, we gonna drop that column

df.drop('id',axis=1,inplace = True)
```

#### 2. url

```python
# This variable is also unrelatable as it just shows the link to the platform website 
df.drop('url',axis=1, inplace = True)
```

```python
print(df.head())
```

#### 3. region

```python
# I am not sure about this variable so we can draw a bar plot to see whether there is any relationship between craigslist region and price
# First try: There are some outliers which make the plot hard to see, we're gonna filter it out (Limitation 100000)
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(data=df[df['price'] < 100000],x='region',y='price')
plt.xlabel('Region')
plt.ylabel('Price')
plt.show()
```

We do not see any pattern from this graph, but we still keep this variable in our final dataset for our nezxt feature engineering process

```python
# Clean the format or the price column
df['region'] = df['region'].str.lower().str.strip()

# Also check if any null value
print(df['region'].isnull().sum())
```

#### 4.region_url

```python
# Because we have the region as an input, we no longer need this column

df.drop('region_url',axis = 1, inplace=True)
```

#### 5.price

```python
# I want to see the distribution of the price
sns.boxplot(x=df['price'])
plt.show()
```

```python
# As our goal for this project is to predict the price of old cars, we're gonna target our vehicles under $200K and of course, greater than $0

df = df[(df['price'] < 200000) & (df['price'] > 0)]

# Draw the distribution again 
sns.histplot(df['price'], bins = 50, kde=True)
plt.show()
```

```python
# Check null value
print(df['price'].isnull().sum())
```

#### 6. year

```python
# We would check the null value first

print(df['year'].isnull().sum())

# We can check the relationship between year and price
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(data=df,x='year',y='price')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```

```python
# Filter out the column that is not null
df = df[df['year'].notna()]
```

```python
# It could be more effective if we use that variable as year_to_now
df["year_to_now"] = 2025 - df['year']
```

#### 7. manufacturer

```python
# We obviously know that manufacturer take a very important role for deciding the price of the car, just need the format it

df['manufacturer'] = df['manufacturer'].str.lower().str.strip()
```

```python
print(df['manufacturer'].value_counts())
print(df['manufacturer'].nunique())
```

```python
print(df['manufacturer'].isnull().sum())
```

```python
df = df[df['manufacturer'].notna()]
print(len(df))
```

#### 8.model

```python
## 8.model
df['model'] = df['model'].str.lower().str.strip()
print(df['model'].nunique())
```

```python
## 22558 is a lot, we now print_out the frequency of each model

print(df['model'].value_counts().head(20))
```

```python
print(df['model'].value_counts().tail(20))
```

```python
# We saw that there are so many models that just appear one time, so we're gonna convert all of that type of model into a name called 'other'
model_counts = df['model'].value_counts()

df = df[df['model'].notna()]

df['model'] = df['model'].map(lambda x: 'other' if model_counts[x] == 1 else x)
```

```python
# Print out the number of unique value in column 'model' and the length of the dataset 

print(df['model'].nunique())
print(len(df))
```

#### 9. condition

```python
# It is hard to say vehicle condition without having everyone on person instead of self describing from the seller. We will remove this column
df.drop('condition',axis = 1, inplace = True)
```

#### 10. cylinders

```python
# First, reformat that words in each cell

df['cylinders'] = df['cylinders'].str.lower().str.strip()

df['cylinders'] = df['cylinders'].fillna('other')

print(df['cylinders'].unique())
```

#### 11.fuel

```python
# Reformat the word

df['fuel'] = df['fuel'].str.lower().str.strip()

print(df['fuel'].isnull().sum())
```

```python
# Fill null value

df['fuel'] = df['fuel'].fillna('other')

print(df['fuel'].unique())
```

#### 12.odometer

```python
# Investigate the distribution

sns.boxplot(df['odometer'])
plt.show()
```

```python
# Look at the plot we can delete limit out scope by setting this number under 4000000
df = df[df['odometer'] < 4000000]

print(len(df))

sns.boxplot(df['odometer'])
plt.show()
```

#### 13. title_status

```python
# Reformat the word

df['title_status'] = df['title_status'].str.lower().str.strip()

print(df['title_status'].unique())

df = df[df['title_status'].notna()]

print(len(df))
```

#### 14. transmission

```python
# Same pattern
df['transmission'] = df['transmission'].str.lower().str.strip()

print(df['transmission'].unique())

df = df[df['transmission'].notna()]

print(len(df))
```

#### 15. VIN

```python
# We do not need the ID of the car
df.drop('VIN',axis=1,inplace=True)
```

#### 16. drive

```python
df['drive'] = df['drive'].str.lower().str.strip()

print(df['drive'].unique())
```

#### 17. size

```python
df['size'] = df['size'].str.lower().str.strip()

print(df['size'].unique())
```

```python
# Fill null value
df['size'] = df['size'].fillna('other')
```

#### 18. type

```python
df['type'] = df['type'].str.lower().str.strip()

print(df['type'].unique())
```

```python
# We can drop any null value that align with the type

df = df[df['type'].notna()]

print(len(df))
```

```python
sns.barplot(x='type',y='price',data=df)
plt.show()
```

#### 19.color

```python
df = df.rename(columns = {'paint_color': 'color'})
df['color'] = df['color'].str.lower().str.strip()

df['color'] = df['color'].fillna('other')

print(df['color'].unique())

print(len(df))
```

#### 20. image_url
#### 21. description
#### 22. county

```python
# All of those 3 variable do not relate to the car price

df.drop('image_url',axis=1,inplace=True)
df.drop('description',axis=1,inplace=True)
df.drop('county',axis=1,inplace=True)
```

#### 23. state

```python
# Reformat it first

df['state'] = df['state'].str.lower().str.strip()

print(df['state'].nunique())
print(df['state'].isnull().sum())
```

#### 24. lat
#### 25. long
#### 26. posting_date

```python
# Those 3 variables are also not related to cars price, so we're gonna drop it

df.drop('lat',axis=1,inplace=True)
df.drop('long',axis=1,inplace=True)
df.drop('posting_date',axis=1,inplace=True)
```
